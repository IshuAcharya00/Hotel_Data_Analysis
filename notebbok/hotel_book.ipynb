```python
def detect_outliers_iqr(df,adr):
    Q1 = df[adr].quantile(0.25)
    Q3 = df[adr].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[adr] < lower_bound) | (df[adr] > upper_bound)]
    return outliers,Q1,Q3,IQR,lower_bound,upper_bound

outliers, Q1, Q3, IQR, lower_bound, upper_bound = detect_outliers_iqr(df, 'adr')

print("Q1:", Q1)
print("Q3:", Q3)
print('lower_bound:', lower_bound)
print('upper_bound:', upper_bound)

print("outliers detected using IQR method:",outliers)
print("NUmber of outliers detected :",len(outliers))
```

```python
# Cancellation Rate by Hotel Type
hotel_cancel_rate = df.groupby('hotel')['is_canceled'].mean().reset_index()

sns.barplot(data=hotel_cancel_rate, x='hotel', y='is_canceled')
plt.title("Cancellation Rate by Hotel Type")
plt.ylabel("Cancellation Rate")
plt.show()
```

```python
# Cancellation Rate by Customer Type
customer_cancel_rate = df.groupby('customer_type')['is_canceled'].mean().reset_index()
sns.barplot(data=customer_cancel_rate, x='customer_type', y='is_canceled')
plt.title("Cancellation Rate by Customer Type") 
plt.ylabel("Cancellation Rate")
plt.show()
print(customer_cancel_rate)
```

```python
#  Cancellation Rate by Season
def Get_season (month):
    if month in ['December','January','February']:
        return 'Winter'
    elif month in ['March','April','May']:
        return 'Spring'
    elif month in ['June','July','August']:
        return 'Summer'
    else:
        return 'Autumn'
    
season_cancel_rate = (
    df.groupby(df['arrival_date_month'].apply(Get_season))['is_canceled']
    .mean()
    .reset_index()
    .rename(columns={'arrival_date_month': 'season'})
)

sns.barplot(data=season_cancel_rate, x='season', y='is_canceled')
plt.title("Cancellation Rate by Customer Type") 
plt.ylabel("Cancellation Rate")
plt.show()
print(season_cancel_rate)
```

```python
# •	Lead time vs cancellation scatter plots
sns.scatterplot(data=df,x='lead_time',y='is_canceled',hue='hotel')
plt.title("Sctterplot of Lead Time vs Cancellation")
plt.xlabel("Lead time")
plt.ylabel("Cancellation Rate")
plt.show()
```

```python
# Revenue Analysis
# revenue impact of Cancellations
revenue = df['adr'] * (df['stays_in_weekend_nights'] + df['stays_in_week_nights'])
# above line calculates the revenue for each booking

revenue_impact = revenue.groupby(df['is_canceled']).sum()
print("Revenue Impact of Cancellations:",revenue_impact)

sns.barplot(x=revenue_impact.index, y=revenue_impact.values)
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.ylabel("Total Revenue")
plt.title("Revenue Impact of Cancellations")
plt.show()
```

```python
# Group by market segment and calculate total revenue

segment_revenue = (
    (df['adr'] * (df['stays_in_weekend_nights'] + df['stays_in_week_nights']))
    .groupby(df['market_segment'])
    .sum()
    .sort_values(ascending=False)
)


print("Market segment revenue:\n", segment_revenue)

# Bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=segment_revenue.index, y=segment_revenue.values)
plt.ylabel('Total Revenue')
plt.title('Revenue by Market Segment')
plt.show()
```

```python

# Create ADR bins temporarily
adr_range = pd.cut(df['adr'], bins=[0, 50, 100, 150, 200, 300, 500], include_lowest=True)

# Count bookings per ADR range
adr_demand = adr_range.value_counts().sort_index()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=adr_demand.index.astype(str), y=adr_demand.values)
plt.xlabel('ADR Range')
plt.ylabel('Number of Bookings')
plt.title('Demand by ADR (Pricing Strategy)')
plt.xticks(rotation=45)
plt.show()
```

```python
cancellation_by_customer_type = df.groupby('customer_type')['is_canceled'].mean().sort_values(ascending=False)

# Plot
sns.barplot(x=cancellation_by_customer_type.index, y=cancellation_by_customer_type.values)
plt.ylabel('Cancellation Rate')
plt.title('Cancellation Rate by Customer Type')
plt.show()
```

```python
# Temporarily bin lead_time
lead_bins = pd.cut(df['lead_time'], bins=[0, 20, 40, 60, 80, 100, 150, 200, 300, 400, 500])

# Calculate cancellation rate per bin
lead_time_cancellation = df.groupby(lead_bins)['is_canceled'].mean()

# Plot
sns.lineplot(x=lead_time_cancellation.index.astype(str), y=lead_time_cancellation.values, marker='o')
plt.xticks(rotation=45)
plt.ylabel("Cancellation Rate")
plt.xlabel("Lead Time Range (Days)")
plt.title("Cancellation Rate vs Lead Time")
plt.show()
```

```python
# Average number of special requests for canceled vs non-canceled bookings
sns.boxplot(x='is_canceled', y='total_of_special_requests', data=df)
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.ylabel('Total Special Requests')
plt.title('Special Requests vs Cancellation')
plt.show()
```

```python
# Which months have the highest booking demand?
monthly_demand = df['arrival_date_month'].value_counts().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

# Plot
sns.barplot(x=monthly_demand.index, y=monthly_demand.values)
plt.xticks(rotation=45)
plt.ylabel("Number of Bookings")
plt.title("Booking Demand by Month")
plt.show()
```

```python
# How do booking patterns differ between hotel types?
# Count bookings per hotel per month
booking_patterns = df.groupby(['arrival_date_month', 'hotel']).size().unstack()

# Reorder months
booking_patterns = booking_patterns.reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

# Plot
booking_patterns.plot(kind='bar', figsize=(12,6))
plt.ylabel("Number of Bookings")
plt.title("Monthly Booking Patterns by Hotel Type")
plt.xticks(rotation=45)
plt.legend(title='Hotel')
plt.tight_layout()
plt.show()
```
